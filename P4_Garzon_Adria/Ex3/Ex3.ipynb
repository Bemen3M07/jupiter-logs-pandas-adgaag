```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el CSV desde un archivo
df = pd.read_csv("notes_alumnes.csv")

# Crear un gráfico de barras de las notas por alumno en la columna 'M01'
plt.figure(figsize=(10, 6))
plt.bar(df['Alumne'], df['M01'], color='skyblue')

# Títulos y etiquetas
plt.title('Notas de los alumnos en M01')
plt.xlabel('Alumno')
plt.ylabel('Nota en M01')
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.tight_layout() 
plt.show()
```

```python
3. Funcionalitats de ReportLab

Funcionalitat 1: Creació de documents PDF
ReportLab permet crear documents PDF personalitzats. Aquesta funcionalitat és útil per generar informes professionals amb taules, gràfics i text.
Exemple de codi:
```

```python
from reportlab.pdfgen import canvas

# Crear un lienzo para el PDF
pdf = canvas.Canvas("exemple.pdf")

# Dibujar una línea
pdf.line(100, 720, 400, 720)

# Agregar texto
pdf.drawString(100, 750, "¡Hola!")

# Dibujar una línea
pdf.line(100, 720, 400, 720)

# Guardar el archivo PDF
pdf.save()

print("El archiuPDF 'exemple.pdf' se ha generat.")
```

```python

Funcionalitat 2: Inserció de gràfics en PDF
ReportLab permet afegir gràfics en documents PDF generats. Aquesta funció és útil per crear informes detallats amb visualitzacions.
Exemple de codi:
```

```python
# Descarregar el gràfic de Matplotlib com una imatge
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.savefig('grafico.png')

# Afegir la imatge al PDF
pdf.drawImage('grafico.png', 100, 500, width=400, height=300)
pdf.save()
```

```python

Funcionalitat 3: Personalització de documents
ReportLab permet personalitzar la tipografia, dissenys de pàgines i altres elements visuals dels documents generats.
Exemple de codi:
```

```python
# Canviar la font i la mida
pdf.setFont("Helvetica-Bold", 12)
pdf.drawString(100, 700, "Aquest text utilitza una font personalitzada.")
pdf.save()
```
