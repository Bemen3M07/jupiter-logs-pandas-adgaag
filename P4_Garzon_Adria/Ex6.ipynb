```python
import pandas as pd
# Importar de la llibreria Matplotlib perquè és una eina potent per a la creació de gràfics i visualitzacions en Python,
# permetent representar de manera clara i intuïtiva les dades, la qual cosa és ideal per a visualitzar les notes dels alumnes 
# i realitzar anàlisis gràfiques de les mitjanes, distribucions i percentatges.
import matplotlib.pyplot as plt

# Intentem carregar el fitxer CSV amb les notes dels alumnes
try:
    df = pd.read_csv("notes_alumnes.csv")  # Llegim el fitxer CSV i el carreguem en un DataFrame
    print("Fitxer carregat correctament!")  # Si el fitxer es carrega bé, mostrem aquest missatge
except FileNotFoundError:
    # Si el fitxer no es troba, mostrem un missatge d'error i sortim del programa
    print("Error: No s'ha trobat el fitxer 'notes_alumnes.csv'. Assegura't que el fitxer estigui al directori correcte.")
    exit() 

# Calcular la mitjana de nota final de cada alumne, fent la mitjana de les notes de les assignatures M01, M04, M03 i M05
df["Mitjana"] = df[["M01", "M04", "M03", "M05"]].mean(axis=1)

# 1. Gràfic de les mitjanes per alumne (diagrama de barres)
plt.figure(figsize=(10, 6))  # Definim la mida del gràfic
plt.bar(df["Alumne"], df["Mitjana"], color='skyblue')  # Creem el gràfic de barres amb les mitjanes
plt.xlabel("Alumnes")  # Títol de l'eix X
plt.ylabel("Mitjana de les notes")  # Títol de l'eix Y
plt.title("Mitjana de les notes per alumne")  # Títol del gràfic
plt.xticks(rotation=90)  # Rotem les etiquetes de l'eix X per a millorar la visibilitat
plt.tight_layout()  # Ajustem el disseny per evitar que les etiquetes es tallin
plt.show()  # Mostrem el gràfic

# 2. Gràfic de les mitjanes per assignatura (diagrama de barres)
assignatures = ["M01", "M04", "M03", "M05"]  # Llista de les assignatures
mitjanes_assignatures = df[assignatures].mean()  # Calculem la mitjana de les notes per assignatura

plt.figure(figsize=(8, 5))  # Definim la mida del gràfic
plt.bar(assignatures, mitjanes_assignatures, color='lightgreen')  # Creem el gràfic de barres
plt.xlabel("Assignatures")  # Títol de l'eix X
plt.ylabel("Mitjana de les notes")  # Títol de l'eix Y
plt.title("Mitjana de les notes per assignatura")  # Títol del gràfic
plt.tight_layout()  # Ajustem el disseny per evitar que les etiquetes es tallin
plt.show()  # Mostrem el gràfic

# 3. Gràfic de distribució de les notes mitjanes (histograma)
plt.figure(figsize=(8, 5))  # Definim la mida del gràfic
plt.hist(df["Mitjana"], bins=10, color='lightcoral', edgecolor='black')  # Creem el histograma
plt.xlabel("Mitjana de les notes")  # Títol de l'eix X
plt.ylabel("Nombre d'alumnes")  # Títol de l'eix Y
plt.title("Distribució de les mitjanes de les notes")  # Títol del gràfic
plt.tight_layout()  # Ajustem el disseny
plt.show()  # Mostrem el gràfic

# 4. Gràfic de percentatge d'aprovats i suspesos (gràfic de sectors)
aprovats = df[df["Mitjana"] >= 5].shape[0]  # Comptem quants alumnes han aprovat (mitjana ≥ 5)
suspesos = df[df["Mitjana"] < 5].shape[0]  # Comptem quants alumnes han suspès (mitjana < 5)
labels = ["Aprovats", "Suspesos"]  # Etiquetes per al gràfic de sectors
sizes = [aprovats, suspesos]  # Dades de les proporcions d'aprovats i suspesos
colors = ['lightgreen', 'salmon']  # Colors per als diferents sectors
explode = (0.1, 0)  # 'Explodem' el primer sector per destacar-lo (els aprovats)

plt.figure(figsize=(7, 7))  # Definim la mida del gràfic
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)  # Creem el gràfic de sectors
plt.title("Percentatge d'aprovats i suspesos")  # Títol del gràfic
plt.axis('equal')  # Per a garantir que el cercle sigui perfecte
plt.show()  # Mostrem el gràfic

```
