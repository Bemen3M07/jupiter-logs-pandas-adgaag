```python
# Importar de la llibreria panda perquè és una eina potent i eficient per a la manipulació i l'anàlisi de dades tabulars, permetent realitzar càlculs estadístics, 
# filtrar i transformar dades de manera senzilla i ràpida, la qual cosa és ideal per a l'anàlisi de les notes dels alumnes en aquest exercici.
import pandas as pd

# Intentem carregar el fitxer CSV
try:
    df = pd.read_csv("notes_alumnes.csv")
    print("Fitxer carregat correctament!")
except FileNotFoundError:
    print("Error: No s'ha trobat el fitxer 'notes_alumnes.csv'. Assegura't que el fitxer estigui al directori correcte.")
    exit()

# Mostrem les primeres línies del DataFrame per assegurar-nos que la càrrega és correcta
print("\nMostra de les dades carregades:")
print(df.head())  # Mostra les primeres 5 files del DataFrame

# Calcular la mitjana de nota final de cada alumne, el promig de les assignatures M01, M04, M03, M05.
df["Mitjana"] = df[["M01", "M04", "M03", "M05"]].mean(axis=1)

# Mostrar la mitjana de cada alumne
print("\nMitjana de cada alumne:")
print(df[["Alumne", "Mitjana"]])

# Calcular la mitjana de nota final de tots els alumnes, el promig global de tots els alumnes.
mitjana_total = df["Mitjana"].mean()
print("\nMitjana total de tots els alumnes:", mitjana_total)

# Percentatges d'aprovats i suspesos, suposant que aprovar és obtenir una mitjana ≥ 5.
aprovats = df[df["Mitjana"] >= 5].shape[0]
suspesos = df[df["Mitjana"] < 5].shape[0]

percentatge_aprovats = (aprovats / len(df)) * 100
percentatge_suspesos = (suspesos / len(df)) * 100

print(f"\nPercentatge d'aprovats: {percentatge_aprovats:.2f}%")
print(f"Percentatge de suspesos: {percentatge_suspesos:.2f}%")

# Notes més baixes i més altes, per cada assignatura.
notes_baixes = df[["M01", "M04", "M03", "M05"]].min()
notes_altes = df[["M01", "M04", "M03", "M05"]].max()

print("\nNotes més baixes per assignatura:")
print(notes_baixes)

print("\nNotes més altes per assignatura:")
print(notes_altes)

# Estadístiques descriptives de les notes, màxim, mínim, mitjana, desviació estàndard.
estadistiques = df[["M01", "M04", "M03", "M05"]].describe()
print("\nEstadístiques descriptives de les notes:")
print(estadistiques)
```
